```python
df = pd.read_csv("data.csv", skiprows = 8, parse_dates=["time"],
                 names = ["time" ,"z", "y", "x", "battery", "celsius", "eda", "event"])
df.drop(["battery", "event"], axis = 1, inplace = True)
df.head(5)
```
